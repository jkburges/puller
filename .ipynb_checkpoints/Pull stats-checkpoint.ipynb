# graph of pulls vs time (over last year?)

```python
import csv
import dateutil.parser
import matplotlib.pyplot as plt

with open('closed_pulls.csv') as closed_pulls_file:
    pull_reader = csv.DictReader(closed_pulls_file, delimiter = ',')
    pulls = []
    for row in pull_reader:
        pulls.append(row)

sorted_pulls = sorted(pulls, key = lambda pull: pull['created_at']) 

count = []
created_at_as_date = []

for index, pull in enumerate(sorted_pulls):
    count.append(index)
    created_at_as_date.append(dateutil.parser.parse(pull['created_at']))
        
%matplotlib inline
plt.figure(figsize=(10,6))
plt.xlabel('date')
plt.ylabel('pull requests')
plt.plot(created_at_as_date, count)
plt.show()
```

## percentage in each project (pie chart)

```python
pull_count_by_project = {}

for pull in pulls:
    if not (pull['project_name'].strip() in pull_count_by_project):
        pull_count_by_project[pull['project_name'].strip()] = 0
        
    pull_count_by_project[pull['project_name'].strip()] += 1


import operator
sorted_pull_count = sorted(pull_count_by_project.items(), key=operator.itemgetter(1))    

threshold = 100
others_count = 0
for label, count in sorted_pull_count:
    if (count < threshold):
        others_count += count

sorted_pull_count = [x for x in sorted_pull_count if not x[1] < threshold]
sorted_pull_count.append(('others', others_count))

labels, pull_counts = zip(*sorted_pull_count)

%matplotlib inline

plt.pie(pull_counts, labels=labels, shadow=True, startangle=90, autopct='%1.1f%%')
plt.show()
    
```

pull age (distribution)

number of comments (distribution)

```python
import numpy as np

merged_pulls = [pull for pull in pulls if len(pull['merged_at'].strip()) > 0]

ages = []
for pull in merged_pulls:
    start_date = dateutil.parser.parse(pull['created_at'])
    end_date = dateutil.parser.parse(pull['merged_at'])
#    print start_date, end_date, pull['merged_at']
    ages.append((end_date - start_date).total_seconds() / 3600)
 

%matplotlib inline

plt.figure(figsize=(10,6))
plt.xlabel('age (hours)')
plt.ylabel('pulls requests')
plt.hist(ages, bins=np.logspace(0.01, 4, 40))
plt.gca().set_xscale("log")
plt.show()

        
```
